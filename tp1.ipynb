# B. Manipulation d’un jeu de données

#### 1- chargement de iris

```python
iris = datasets.load_iris()
```

#### 2- Affichages les données, les noms des variables et le nom des classes

###### Affichage des données 

```python
print("Les données sont : ")
print(iris.data)
```

##### Affichage des noms de variable

```python
print("Les noms de variable sont : ")
print(iris.feature_names)
```

##### Affichage des noms de classe

```python
print("Les noms de classe sont : ")
print(iris.target_names)
```

#### 3- affichage de nom de classe pour chaque donnée

```python
for i in range(len(iris.data)):
    print(f"La donnée numero {i+1} : {iris.data[i]} est de classe: {iris.target_names[iris.target[i]]}")
```

#### 4- Affichage de la moyenne, l’ecart-type, le min et le max pour chaque variable.

```python
for i in range(iris.data.shape[1]):
    print(f"La variable {iris.feature_names[i]} est de moyenne :{iris.data.mean(0)[i]} ,ecart_type :{iris.data.std(0)[i]} ,min :{iris.data.min(0)[i]},max :{iris.data.max(0)[i]}")
    print("\n")
```

#### 5- affichage du nombre de données, le nombre de variables et le nombre de classes.

```python
print(f"Nombre de données : {iris.data.shape[0]}")
print(f"Nombre de variables : {iris.data.shape[1]}")
print(f"Nombre de classes : {iris.target_names.size}")
```

# C. Téléchargement et importation de données

#### 1- importation de mnist

```python
mnist = datasets.fetch_openml('mnist_784', version=1, parser='auto')
```

#### 2- Affichage detaillé

```python
print(f"La matrice de donnees : {mnist.data}")
print(f"Nombre de données : {mnist.data.shape[0]}")
print(f"Nombre de variables : {mnist.data.shape[1]}")
for i in range(len(mnist.data)):
    print(f"La donnée numero {i+1} est de classe numero : {mnist.target[i]}")
print(f"La moyenne :{mnist.data.mean(0)}")
print(f"L\'ecart_type :{mnist.data.std(0)}")
print(f"Le min :{mnist.data.min(0)}")
print(f"La max :{mnist.data.max(0)}")
print(f'le nombre de classe est : {np.unique(mnist.target).size}')
```

# D. Génération de données et affichage

#### 2- Génération de 1000 données de deux variables réparties en 4 groupes.

```python
X, y = datasets.make_blobs(n_samples=1000, centers=4, n_features=2)
```

#### 3- visualisation

```python
plt.Figure()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Visualisation des données générées avec make_blobs")
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.show()
```

#### 4- Génération et concatenation et visualisation 

```python
# Générer 100 données réparties en 2 groupes
X1, y1 = datasets.make_blobs(n_samples=100, centers=2, n_features=2)

# Générer 500 données réparties en 3 groupes
X2, y2 = datasets.make_blobs(n_samples=500, centers=3, n_features=2)

XX = np.vstack((X1, X2))
yy = np.hstack((y1, y2+2))
 

plt.Figure()
plt.scatter(XX[:, 0], XX[:, 1], c=yy)
plt.title("Visualisation des données ")
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.show()
```
